1. Literacy rates for 1951, 1961 and 1971 Censuses relate to population aged five years and above. The rates for the 1981, 1991 and 2001
Censuses relate to the population aged seven years and above. The literacy rate for 1951 in case of West Bengal relates to Total
population including 0-4 age group. Literacy rate for 1951 in respect of Chhatisgarg, Madhya Pradesh and Manipur are based on sample
population.

2. India and Manipur figures exclude those of the three sub-divisions viz. Mao Maram, Paomata and Purul of Senapati district of Manipur
as census results of 2001 in these three sub-divisions were cancelled due to technical and administrative reasons.

3. N.A. - Not available 

4. No census was carried out in Assam during 1981 and in Jammu & Kashmir during 1991.

5. Created in 2001. Uttaranchal Pradesh, Jharkhand and Chattisgarh for 1981 and 1991 are included under Uttar Pradesh, Bihar and Madhya Pradesh respectively.

7. Daman and Diu came into existence on30th May, 1987.

8. State Fromation of Haryana & Panjab Date : 1 November 1966.

# * Req_3 : Area of each state *

Indian States by Area, Ref: https://www.findeasy.in/top-indian-states-by-area/

```python
#Read data from specified URL

area_df1 = pd.read_html("https://www.findeasy.in/top-indian-states-by-area/",header=0)
df31 = pd.DataFrame(area_df1[0])
df31
```

```python
df31.sort_values(by='States/UT',inplace=True)
df31.drop(columns='Rank',inplace=True)
df31.reset_index(inplace=True,drop=True)
df31
```

Indian UT by Area, Ref: https://www.findeasy.in/top-indian-states-by-area/

```python
#Read data from specified URL

area_df2 = pd.read_html("https://www.findeasy.in/top-indian-states-by-area/",header=0)
df32 = pd.DataFrame(area_df2[1])
df32
```

```python
df32.sort_values(by='Union Territories',inplace=True)
df32.drop(columns='Rank',inplace=True)
df32.rename(columns = {'Union Territories':'States/UT'}, inplace = True)
df32.replace('National Capital Territory of Delhi','Delhi',inplace=True)
df32.reset_index(inplace=True,drop=True)
df32
```

```python
df3 = pd.concat([df31,df32],axis=0)
df3.sort_values(by='States/UT',inplace=True)
df3.drop(columns='Density* (p/km2)',inplace=True)
df3.reset_index(inplace=True,drop=True)
df3
```

```python
Indian_StateUT_Area_File = df3
Indian_StateUT_Area_File.to_csv('Indian_StateUT_Area.csv')
```
